```python
#Gradient descent algorithm
```

```python
def LNfunc(X,theta):
    return theta[0]+theta[1]*X

def error(X,Y,theta):
    m=X.shape[0]
    err=0
    for i in range(m):
        hx=LNfunc(X[i],theta)
        err+=(hx-Y[i])**2
    return 1/m*err

def Grad(X,Y,theta):
    grad=np.zeros((2,))
    m=X.shape[0]
    for i in range(m):
        hx=LNfunc(X[i],theta)
        grad[0]+=(hx-Y[i])
        grad[1]+=(hx-Y[i])*X[i]
    return grad

def GradientDescent(X,Y,learning_rate=0.001):
    theta=np.zeros((2,))
    
    grad=Grad(X,Y,theta)
    itr=0
    error_list=[]
    while(itr<=100):
        e=error(X,Y,theta)
        error_list.append(e)
        theta[0]=theta[0]-learning_rate*grad[0]
        theta[1]=theta[1]-learning_rate*grad[1]
        itr+=1
        ind=np.argmin(error_list)
    
    return ind
        
def GradientDescent1(X,Y,learning_rate=0.001):
    theta=np.zeros((2,))
    
    grad=Grad(X,Y,theta)
    itr=0
    ind=GradientDescent(X,Y,learning_rate=0.001)
    error_list=[]
    while(itr<=ind):
        e=error(X,Y,theta)
        error_list.append(e)
        theta[0]=theta[0]-learning_rate*grad[0]
        theta[1]=theta[1]-learning_rate*grad[1]
        itr+=1
        
    
    return theta,error_list      
    
```

```python
final_theta,error_list=GradientDescent1(X,Y)
```
