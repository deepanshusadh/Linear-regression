```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import make_regression
```

```python
# Generate DATA set
X,Y=make_regression(n_samples=400,n_features=1,n_informative=1,noise=1.8,random_state=11)
```

```python
print(X.shape,Y.shape)
```

```python
Y=Y.reshape((-1,1))
print(Y.shape)
```

```python
X=(np.mean(X)-X)/np.std(X)
```

```python
plt.plot(X,Y)

plt.scatter(X,Y,color='orange')
```

```python
plt.scatter(X,Y,color='orange')
```

```python
X_one=np.ones((X.shape[0],1))
X_one.shape
```

```python
X_br=np.hstack((X,X_one))
X_br.shape
```

```python
P=X_br.T
P.shape
X_br.shape
```

```python
def predict(X,theta):
    return np.dot(X,theta)

def CFS(X,Y):
    Y=np.mat(Y)
    theta=np.linalg.pinv(np.dot(X.T,X))*np.dot(X.T,Y)
    return theta
```

```python
T=CFS(X_br,Y)
print(T)
```

```python
Pred=predict(X_br,T)
```

```python
plt.scatter(X,Y)
plt.plot(X,Pred,color='orange')
```
